```python
import pandas as pd

df = pd.read_csv("Titanic-Dataset.csv")

df.head()
```

```python
df.isnull().sum()
```

```python
# Age ke missing values ko average age se fill karna
df["Age"] = df["Age"].fillna(df["Age"].mean())

# Embarked ke missing values ko most common value se fill karna
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Cabin column remove karna (bahut zyada missing values hain)
df = df.drop("Cabin", axis=1)

print("Data Cleaned Successfully")
```

```python
import pandas as pd

df = pd.read_csv("Titanic-Dataset.csv")

# Age ke missing values ko average age se fill karna
df["Age"] = df["Age"].fillna(df["Age"].mean())

# Embarked ke missing values ko most common value se fill karna
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Cabin column remove karna
df = df.drop("Cabin", axis=1)

print("Data Loaded")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x="Sex", hue="Survived", data=df)

plt.title("Survival by Gender")

plt.savefig("chart1_survival_by_gender.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Age"], bins=30)

plt.title("Age Distribution")

plt.savefig("chart2_age_distribution.png")

plt.show()
```
